```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
data = pd.read_csv('test.csv')
data.dropna(inplace=True)
```

```python
plt.plot(np.array(data['gram'],dtype=np.float32))
```

```python
voltage = np.array(data['voltage'], dtype=np.float32)
param = linregress(data['az'], voltage)
plt.scatter(data['az'], voltage)
plt.plot(data['az'], param[0] * data['az'] + param[1], color='red')
plt.show()
plt.scatter(data['az'], voltage - param[0] * data['az'] - param[1])
```
